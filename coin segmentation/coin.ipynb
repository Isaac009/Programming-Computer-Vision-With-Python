```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def main_processor(circles):
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # draw the outer circle
        cv2.circle(img,(i[0],i[1]),i[2],(0,0,255),10)
        # draw the center of the circle
        cv2.circle(img,(i[0],i[1]),2,(0,255,0),10)

    plt.rcParams['figure.figsize'] = (16,10)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Final result")
    plt.show()

    print("There are : " + str(circles.shape[1]) + " coins")
```

```python
img = cv2.imread('IMG_8253.jpg')
x = cv2.medianBlur(img,5)
cimg = cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)
blur = cv2.medianBlur(cimg, 25, 0)

circles = cv2.HoughCircles(blur,cv2.HOUGH_GRADIENT,1,80,param1=50,param2=42,minRadius=90,maxRadius=0)

main_processor(circles)
```
