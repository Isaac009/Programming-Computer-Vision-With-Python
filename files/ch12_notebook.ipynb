```python
from PIL import Image
from pylab import *
# import matplotlib.pyplot as pl
# read image to array
im = array(Image.open('data/white.jpg'))
# plot the image
imshow(im)
x = [150,250]
y = [90,90]
plot(x,y,'r*')
plot(x[:],y[:],'g')
# plot the points with red star-markers
plot(x,y,'r*')
# line plot connecting the first two points
plot(x[:2],y[:2],'b')
# add title and show the plot
title('Plotting: "white.jpg"')
show()
#plot(x,y) # default blue solid line
#plot(x,y,r*) # red star-markers
#plot(x,y,go-) # green line with circle-markers
#plot(x,y,ks:) # black dotted line with square-markers
```

```python
# read image to array
im = array(Image.open('data/white.jpg').convert('L'))
# create a new figure
figure()
# dont use colors

gray()
imshow(im)
figure()
# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')
show()
```
