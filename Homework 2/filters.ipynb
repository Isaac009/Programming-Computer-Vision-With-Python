```python
import matplotlib.pyplot as plt
from PIL import Image
from pylab import *

import cv2
import numpy as np

%matplotlib inline
```

```python
# image = matImage.imread('./data/images.jpg')
image = array(Image.open('./data/mdada.jpg'))
gray_image = array(Image.open('./data/mdada.jpg').convert('L'))
plt.imshow(gray_image, cmap='gray')
```

```python
gblur = np.array([[1,2,1],
                 [2,4,2],
                 [1,2,1]])/16

box_blur = np.array([[1,1,1],
                 [1,1,1],
                 [1,1,1]])/9

sobel_y = np.array([[-1,-2,-1],
                   [-2,0,2],
                   [-1,2,1]])

edge_detection1 = np.array([[1,0,-1],
                           [0,0,0],
                           [-1,0,1]])

edge_detection2 = np.array([[0,1,0],
                           [1,-4,1],
                           [0,1,0]])

edge_detection3 = np.array([[-1,-1,-1],
                           [-1,8,-1],
                           [-1,-1,-1]])

sharpen =  np.array([[0,-1,0],
                    [-1,5,-1],
                    [0,-1,0]])
```

```python
filtered_image_box_blur = cv2.filter2D(gray_image, -1, box_blur)

filtered_image_gblur = cv2.filter2D(gray_image, -1, gblur)

fil_img_y = cv2.filter2D(filtered_image_box_blur, -1, sobel_y)

gray()
fig, axs = subplots(1, 4, figsize=(28, 28))
axs[0].imshow(gray_image)
axs[0].axis('off')
axs[1].imshow(filtered_image_box_blur)
axs[1].axis('off')
axs[2].imshow(filtered_image_gblur)
axs[2].axis('off')
axs[3].imshow(fil_img_y)
axs[3].axis('off')
show()
```

```python
filtered_image_gblur = cv2.filter2D(gray_image, -1, gblur)

fil_img_sharpen = cv2.filter2D(gray_image, -1, sharpen)

gray()
fig, axs = subplots(1, 3, figsize=(28, 28))
axs[0].imshow(gray_image)
axs[0].axis('off')
axs[1].imshow(filtered_image_gblur)
axs[1].axis('off')
axs[2].imshow(fil_img_sharpen)
axs[2].axis('off')
show()
```

```python
filtered_image_det1 = cv2.filter2D(gray_image, -1, edge_detection1)

filtered_image_det2 = cv2.filter2D(gray_image, -1, edge_detection2)

filtered_image_det3 = cv2.filter2D(gray_image, -1, edge_detection3)

gray()
fig, axs = subplots(1, 4, figsize=(28, 28))
axs[0].imshow(gray_image)
axs[0].axis('off')
axs[1].imshow(filtered_image_det1)
axs[1].axis('off')
axs[2].imshow(filtered_image_det2)
axs[2].axis('off')
axs[3].imshow(filtered_image_det3)
axs[3].axis('off')
show()
```
