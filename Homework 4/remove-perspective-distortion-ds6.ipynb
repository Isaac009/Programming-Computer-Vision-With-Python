```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.imread('images/dsu6.jpg')
```

```python
cv2.circle(cap, (110, 150), 5, (255, 0, 0), -1)
cv2.circle(cap, (720, 240), 5, (255, 0, 0), -1)
cv2.circle(cap, (110, 576), 5, (255, 0, 0), -1)
cv2.circle(cap, (720, 535), 5, (255, 0, 0), -1)
plt.imshow(cap)
```

```python
pts1 = np.float32([[110, 150], [700, 240], [110, 572], [715, 532]])
pts2 = np.float32([[0, 0], [700, 0], [0, 470], [700, 470]])
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(cap, matrix, (700, 470))
```

```python
plt.imshow(result)
```
