```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.imread('images/dsu9.jpg')
```

```python
cv2.circle(cap, (395, 215), 5, (255, 0, 0), -1)
cv2.circle(cap, (655, 220), 5, (255, 0, 0), -1)
cv2.circle(cap, (395, 600), 5, (255, 0, 0), -1)
cv2.circle(cap, (645, 580), 5, (255, 0, 0), -1)
plt.imshow(cap)
```

```python
pts1 = np.float32([[400, 210], [650, 220], [400, 595], [635, 570]])
pts2 = np.float32([[0, 0], [500, 0], [0, 500], [500, 500]])
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(cap, matrix, (500, 500))
```

```python
plt.imshow(result)
```
