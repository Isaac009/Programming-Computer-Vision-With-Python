```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.imread('images/dsu2.jpg')
```

```python
cv2.circle(cap, (178, 240), 5, (255, 0, 0), -1)
cv2.circle(cap, (814, 226), 5, (255, 0, 0), -1)
cv2.circle(cap, (182, 500), 5, (255, 0, 0), -1)
cv2.circle(cap, (814, 510), 5, (255, 0, 0), -1)
plt.imshow(cap)
```

```python
pts1 = np.float32([[178, 240], [814, 226], [182, 500], [814, 510]])
pts2 = np.float32([[0, 0], [900, 0], [0, 500], [900, 500]])
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(cap, matrix, (900, 500))
```

```python
plt.imshow(result)
```
