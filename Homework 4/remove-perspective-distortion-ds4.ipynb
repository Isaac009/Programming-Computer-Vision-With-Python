```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cap = cv2.imread('images/dsu7.jpg')
```

```python
cv2.circle(cap, (110, 230), 5, (255, 0, 0), -1)
cv2.circle(cap, (915, 170), 5, (255, 0, 0), -1)
cv2.circle(cap, (110, 513), 5, (255, 0, 0), -1)
cv2.circle(cap, (905, 530), 5, (255, 0, 0), -1)
plt.imshow(cap)
```

```python
pts1 = np.float32([[186, 240], [910,174], [175, 495], [900, 525]])
pts2 = np.float32([[0, 0], [900, 0], [0, 485], [900, 485]])
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(cap, matrix, (900, 475))
```

```python
plt.imshow(result)
```
