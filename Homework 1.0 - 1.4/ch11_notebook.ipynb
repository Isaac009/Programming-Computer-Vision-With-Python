```python
import os
filelist = os.listdir('data')
for infile in filelist:
    outfile = os.path.splitext(infile)[0] + ".jpg"
    print(outfile)
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print("cannot convert", infile)
```

```python
box = (0,0,80,80)
region = pil_im.crop(box)
pl.figure()
pl.imshow(region)
pl.gray()
pl.axis('off')
pl.show()
```

```python
def showimage(src):
    pl.figure()
    pl.imshow(src)
    pl.axis('off')
    pl.show()
```

```python
showimage(region)
```

```python
region = region.transpose(Image.ROTATE_180)
showimage(region)
```

```python
pil_im.paste(region,box)
showimage(pil_im)
```
